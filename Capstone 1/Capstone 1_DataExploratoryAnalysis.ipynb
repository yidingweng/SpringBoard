```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import csv
import seaborn as sns
```

```python
print(pre90s["Pos"].value_counts())
```

```python
print(_90s["Pos"].value_counts())
```
