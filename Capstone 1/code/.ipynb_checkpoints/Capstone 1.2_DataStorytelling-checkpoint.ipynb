# Data Storytelling

### Main purpose

Basketball players at different positions have very different performance in each category. For example, player in the role of point guard(PG) are the ones with smaller statues, they tend to be the ball handlers most of the time, and often has higher number of assists in the game. And bigger players in the role of center(C) or power forward(Pf) tend to grab more rebounds and blocks since the rim is more accessable for them. 
The main purpose of this step is to examine how these common understandings of basketball player function stand firm in the light of real data, and we may find trend of changes in data attribute in relation to player position and time, which help us to see whether the function of each position has evolved into more specialized and distinctive in its own style: Does PG have more assist? Does center grab even more rebounds? Or whether the functions of different position has become more intermingled and interchangable? Therefore I will divide the cleansed data according to position and according to time. The row data attributes I will look at are:

    PTS - Number of points
    3PA - number of 3-Pointer attempt
    2PA - number of 2-Pointer attempt
    AST - number of assists
    TRB - number of total rebound
    PF - number of personal fouls
    FT - number of free throws
    BLK - number of blocks

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import csv
import seaborn as sns
```

```python
with open('data/Seasons_Stats_cleansed.csv') as csvfile:
    reader1 = csv.reader(csvfile)
data = pd.read_csv('data/Seasons_Stats_cleansed.csv',sep='\t')
```

### Divide the datasets into 5 sets, each is corroesponding to one position

```python
pg = data[data['Pos'] == 'PG']
sg = data[data['Pos'] == 'SG']
sf = data[data['Pos'] == 'SF']
pf = data[data['Pos'] == 'PF']
c = data[data['Pos'] == 'C']
```

### Divide the datasets into 4 sets, each is corroesponding to one time period
    pre 90s - from 1978 to 1990
    _90s - from 1991 to 2000
    _00s - from 2001 to 2010
    _10s - from 2011 to 2017

```python
pre90s = data[~(data['Year']>1989)]
_90s = data[~(data['Year']<1990) & ~(data['Year']>1999)]
_00s = data[~(data['Year']<2000) & ~(data['Year']>2009)]
_10s = data[~(data['Year']<2010)]
```

```python
import matplotlib.style
import matplotlib as mpl
import matplotlib.pyplot as plt
# to change default colormap
plt.rcParams["image.cmap"] = "Set1"
# to change default color cycle
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=plt.cm.Set1.colors)
```

### Use separate scatter plot and line plot to study the general trend of certain attribute number across the time for each position, then use box plot to observe the proportional distribution of data within each position in four decades.

For each box plot, I will arrage the box for each position in the order of
PG, SG, SF, PF, C
which usually also implies height of players in ascanding order. 

### Build scatter plot function 

We will make one scatter plot for each position in the order of PG, SG, SF, PF, C and ALL POSITIONS, which means this function should generate 6 plots at once. Once the function is built, user can input plot title and attribute category to observe the change of attribute across time for indvidual players of each position.

```python
positionStrings = ['PG','SG','SF','PF','C','ALL POSITIONS']
positions = [pg,sg,sf,pf,c,data]
plotOrderX = [0,1,2,0,1,2]
plotOrderY = [0,0,0,1,1,1]

def scatterPlot2x3(plotTitle,plotCategory):
    fig, axarr = plt.subplots(2, 3, figsize=(20, 15))
    fig.suptitle(plotTitle,fontsize=20)
    for i in range(6):
        positions[i].plot(kind='scatter',x='Year',y=plotCategory,ax=axarr[plotOrderY[i]][plotOrderX[i]])
        axarr[plotOrderY[i],plotOrderX[i]].set_title(positionStrings[i])
```

### Build line plot function for averaging subgroup attribute number.

We keep the same grouping as the scatter plot, which is according to player position, but for the line plot the focus will be on the average number of each attribute category. After the function si built, we can observe the general trend of any attribute category throughout time for each position.

```python
def averageLinePlot(plotTitle,plotCategory):
    plt.figure(figsize=(30,20))
    plt.title(plotTitle,fontsize=20)
    for i in range(6):
        bench_by_size = positions[i].groupby('Year')
        means = bench_by_size.mean()
        plt.plot(means[plotCategory],label=positionStrings[i], linestyle='-', marker='o', markersize=5)
    plt.legend(loc='upper right')
    plt.show()
```

### Build box plot function 

For the box plot we change the general grouping from player position to time period, which gives a broader observation of player behavior since each box summarize the distribution of attribute data in 10 years, we can also see how group difference changed throughout different decades.

```python
timePeriodStrings = ['1978 to 1990','1991 to 2000','2001 to 2010','2011 to 2017']
timePeriods = [pre90s,_90s,_00s,_10s]
plotOrderX = [0,1,0,1]
plotOrderY = [0,0,1,1]

def boxPlot2x2(plotTitle,plotCategory):
    fig, axarr = plt.subplots(2, 2, figsize=(30, 20))
    fig.suptitle(plotTitle,fontsize=20)
    for i in range(4):
        sns.boxplot(x="Pos", y=plotCategory, data=timePeriods[i],order=["PG", "SG", "SF","PF","C"],ax=axarr[plotOrderY[i]][plotOrderX[i]])
        axarr[plotOrderY[i],plotOrderX[i]].set_title(timePeriodStrings[i])
```

## PTS (number of points)

#### PTS (number of points) of individual player in each position from 1978-2017

```python
scatterPlot2x3('Number of points for individual player at position of ...','PTS')       
```

As we see from the above 6 Number of points decreases in SG and SF, in other positions stays around the same.
For SG most of players scored 10 to 25 around 1980s, and decreased to 5 to 22 around 2015. For SF, most of players scored 10 to 22 around 1980s, and decreased to 8 to 20 around 2015.

####  Take the average number of points per game of each players in different positions

```python
averageLinePlot('Average number of points of each position','PTS')
```

In this average line plot we continue to observe a drastic decrease in number until 1997, especially in the position of SG and SF. For SG, the average number was between 17 and 18 around 1980, decreased to around 15 around 2000, and the number has remained same. For SF, the average number was between 16 and 18 around 1980 and decreased to 13-15 around and after 2000. Significant increase happens in PG and C after 2000, where we see in both postions the average number of score increased from 13 at 2000, to 15 or 16 after 2015.

#### Observe the number of points for player distribution in each time period

```python
boxPlot2x2('Number of points for player distribution in each time period ...',"PTS")
```

The box plot also suggests that number of points attained by SG and SF was obviously more than other three positions except from 2011 to 2017, especially for SF, its 25% and 75% mark was the highest among all positions before 1990s, but both marks becomes the lowerest after 2010. And if we pay attention to PG and C, both their 25% and 75% were lower than the other positions before 1990, but after 2010 these numbers rised up and catched up to the average.
