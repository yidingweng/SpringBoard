As of v2.0, spaCy supports models trained on more than one language, in this case is English. This is especially useful for named entity recognition. 

```python
nlp = spacy.load('en')
frames = []
for i,row in tqdm_notebook(data2.iterrows()):
    doc = row['content']
    text_id = row['id']
    doc = nlp(str(doc))
    ents = [(e.text, e.start_char, e.end_char, e.label_) 
            for e in doc.ents 
            if len(e.text.strip(' -—')) > 0]
    frame = pd.DataFrame(ents)
    frame['id'] = text_id
    frames.append(frame)
```

```python
npf = pd.concat(frames)
npf.head(10)
```

```python
df1 = pd.DataFrame(np.array(trump_list_Breitbart))
df2 = pd.DataFrame(np.array(trump_list_NewYorkPost))
df3 = pd.DataFrame(np.array(trump_list_NPR))
```

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n
    return x, y
```

```python
# Compute ECDFs
x_sg_Breitbart, y_sg_Breitbart = ecdf(df1.iloc[:,0])
x_sg_NewYorkPost, y_sg_NewYorkPost = ecdf(df2.iloc[:,0])
x_sg_NPR, y_sg_NPR = ecdf(df3.iloc[:,0])

plt.figure(figsize=(30,20))

# Plot the ECDFs
_ = plt.plot(y_sg_Breitbart,x_sg_Breitbart, marker='.', linestyle='none')
_ = plt.plot(y_sg_NewYorkPost,x_sg_NewYorkPost, marker='.', linestyle='none')
_ = plt.plot(y_sg_NPR,x_sg_NPR, marker='.', linestyle='none')

# Set margins
plt.margins(0.02)

# Add axis labels and legend
_ = plt.xlabel('SG points')
_ = plt.ylabel('ECDF')
_ = plt.legend(('Breitbart', 'NewYorkPost', 'NPR'), loc='lower right')

# Show the plot
plt.show()
```

```python
def bootstrap_replicate_1d(data, func):
    return func(np.random.choice(data, size=len(data)))

def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    # Store the replicate in the appropriate index of bs_replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)

    return bs_replicates
```

```python
mean_diff = np.mean(df3.iloc[:,0]) - np.mean(df2.iloc[:,0])

print('difference of means =', mean_diff, 'pts')

# Compute mean of combined data set: combined_mean
combined_mean = np.mean(np.concatenate((df3.iloc[:,0], df2.iloc[:,0])))

# Shift the samples
sg_pre90s_shifted = df3.iloc[:,0] - np.mean(df3.iloc[:,0]) + combined_mean
sg_10s_shifted = df2.iloc[:,0] - np.mean(df2.iloc[:,0]) + combined_mean

# Get bootstrap replicates of shifted data sets
sg_replicates_pre90s = draw_bs_reps(sg_pre90s_shifted, np.mean, 10000)
sg_replicates_10s = draw_bs_reps(sg_10s_shifted, np.mean, 10000)

# Compute replicates of difference of means: bs_diff_replicates
sg_diff_replicates = sg_replicates_pre90s - sg_replicates_10s

# Compute the p-value: p
p = np.sum(sg_diff_replicates >= mean_diff) / len(sg_diff_replicates)

# Print p-value
print('p =', p)
```
